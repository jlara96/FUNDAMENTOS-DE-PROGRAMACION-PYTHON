e) En este apartado vamos a generar 3 grupos con un número de muestras similares, utilizando el metodo "qcut". Etiquetaremos estos tres grupos en la columna "Price_Category", con las etiquetas "low","medium" y "high". Crea esta columna, y muestra cuántas filas hay para cada uno de sus valores, así como el precio medio de cada uno de estos grupos.
**(1 punto)** <span style = "font-family: Courier New; background-color: #f2ae72; color: #000000; padding: 3px;">EI</span>

```python
#Creamos la nueva columna con los datos que contendran con el siguiente código
etiquetas = ["low", "medium", "high"]
elec_car["Price_Category"] = pd.qcut(elec_car["PriceEuro"], 3, etiquetas)
```

```python
bajo = len(elec_car[(elec_car.Price_Category == "low")])
medio = len(elec_car[(elec_car.Price_Category == "medium")])
alto = len(elec_car[(elec_car.Price_Category == "high")])


print(f"el numero de filas que tienen low en la variable Price Cartegory son {bajo}")
print(f"el numero de filas que tienen medium en la variable Price Cartegory son {medio}")
print(f"el numero de filas que tienen high en la variable Price Cartegory son {alto}")
```

```python
# Mostramos los datos agrupados en Price category y mostramos la medio de los precios.
grup_pre = elec_car.groupby("Price_Category")
mean_group_pre = grup_pre.aggregate({"PriceEuro": np.mean})
mean_group_pre
```

(f)  Tras generar esta segmentación, los siguientes puntos que parecen importar a nuestro compañero son la aceleración y la autonomía. Crea un dataframe que contenga solo los vehículos de la "price_Category" "low" y normaliza las columnas "AccelSec" y "Range_Km". Normaliza dichas dos variables utilizando el módulo `MinMaxScaler` de sklearn. Como la Acceleración es mejor como más pequeña es, deberemos invertir los valores obteniendo un 0 para el valor de "AccelSec" mas alto, y un 1 para el "AccelSec" menor. Crea las dos columnas nuevas para los valores normalizados, y una tercera con el producto de los mismos, que llamaremos "Product".
**(1 punto)** <span style = "font-family: Courier New; background-color: #f2ae72; color: #000000; padding: 3px;">EI</span>

```python
# Seleccionamos las filas que contengan low en Price Category y creamos las nuevas variables 
elecpartner = elec_car[(elec_car.Price_Category == "low")]
elecpartner["AccelSecnorm"] = elecpartner["AccelSec"]

elecpartner["Range_kmnorm"] = elecpartner["Range_Km"]
elecpartner["Product"] = ""
```

```python
# Utilizamos la siguiente funcion para normalizar los datos
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range= (0,1)).fit(elecpartner.iloc[:,[15,16]])
scaler.transform(elecpartner.iloc[:,[15,16]])
```

```python
#introducimos los datos normalizados en sus respectivas variables
elecpartner.iloc[:,[15,16]] = scaler.transform(elecpartner.iloc[:,[15,16]])

#Invertimos los valores para adecuarlos a las mediciones del enunciado
elecpartner["AccelSecnorm"] = 1 - elecpartner["AccelSecnorm"]

#introducimos los datos en la variable Product
elecpartner["Product"] = elecpartner["AccelSecnorm"] * elecpartner["Range_kmnorm"]
```
