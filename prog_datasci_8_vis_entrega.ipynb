a) Usando `networkx`, crea un grafo no dirigido con las conexiones declaradas anteriormente. Representa este grafo con `spring_layout`, y con los nombres de las ciudades representadas en cada nodo.  


<span style="font-family: Courier New; background-color: #f2ae72; color: #000000; padding: 3px; ">EI</span>
**(1 punto)**




````python
import networkx as nx

G = nx.Graph() # crear un grafo
#introducimos los nodos
G.add_node('St. Louis')
G.add_node('San Diego')
G.add_node('Chicago')
G.add_node('San Francisco')
G.add_node('Minneapolis')
G.add_node('Portland')
G.add_node('Seattle')
G.add_node('Tulsa')
G.add_node('New York')
G.add_node('Dallas')
G.add_node('Phoenix')
G.add_node('Cleveland')
G.add_node('Denver')
G.add_node('Los Angeles')
G.add_node('Philadelphia')
G.add_node('Atlanta')
G.add_node('Albuquerque')
G.add_node('Philadelphia')

#Introducimos las interacciones
G.add_edge('Seattle', 'Portland')
G.add_edge('Los Angeles', 'Seattle')
G.add_edge('Minneapolis', 'Portland')
G.add_edge('Seattle', 'Portland')
G.add_edge('Albuquerque', 'Atlanta')
G.add_edge('Philadelphia', 'Atlanta')
G.add_edge('Boston', 'Atlanta')
G.add_edges_from([('San Francisco', 'Portland'), ('San Francisco', 'Seattle'),('San Francisco', 'Cleveland')])
G.add_edges_from([('Denver', 'Minneapolis'),('Denver', 'Cleveland')])
G.add_edges_from([('Dallas', 'Cleveland'),  ('Dallas', 'Albuquerque')])
G.add_edges_from([('Miami', 'New York'),   ('Miami', 'Philadelphia'),('Miami', 'Denver')])
G.add_edges_from([('Chicago', 'New York'),   ('Chicago', 'Los Angeles')])
G.add_edges_from([('Phoenix', 'Cleveland'),('Phoenix', 'Denver'),('Phoenix', 'Dallas')])
G.add_edges_from([('Tulsa', 'New York'),  ('Tulsa', 'Dallas')])
G.add_edges_from([('San Diego', 'Chicago'), ('San Diego', 'San Francisco'),('San Diego', 'Minneapolis'),('San Diego', 'Boston'), ('San Diego', 'Portland'), ('San Diego', 'Seattle')])
G.add_edges_from([('St. Louis', 'Miami'), ('St. Louis', 'San Diego'),('St. Louis', 'Chicago')])

````

````python
import matplotlib.pyplot as plt
# Generamos una nueva figura.
plt.figure(1, figsize=(20, 20)) 
# Calculamos las posiciones de los nodos del grafo al plano con el algoritmo
# spring.
graph_pos = nx.spring_layout(G)

# Calculamos el grado de los nodos del grafo.
d = nx.degree(G)

# Mostramos los nodos del grafo, especificando la posición, el tamaño (en función del grado), 
# el color y la transparencia.
nx.draw_networkx_nodes(G, graph_pos, 
                       node_size=8000, 
                       node_color='orange', 
                       alpha=0.3)

# Mostramos las aristas del grafo especificando la posición, 
# el color y la transparencia.
nx.draw_networkx_edges(G, graph_pos, edge_color='blue', alpha=0.3)

# Mostramos los hashtags indicando la fuente y el tamaño.
a = nx.draw_networkx_labels(G, graph_pos, font_size=20, font_family='sans-serif')
````

b) Calcula utilitzando las herramientas de NetworkX todas las posibles rutas entre `Philadelphia` y `San Francisco`, y muéstra la primera (la que cruza menos ciudades) y la última (la que cruza más ciudades). ¿Cuantas rutas aparecen? ¿Cuantas ciudades contiene la más larga ?

**NOTA** Puedes consultar en [shortest_simple_paths](https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.simple_paths.shortest_simple_paths.html) cómo generar caminos entre nodos.

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span>
**(1 punto)**

````python
#Creamos la funcion para encontrar los caminos
caminos = list(nx.shortest_simple_paths(G, 'Philadelphia', 'San Francisco', weight=None))

#Seleccionamos el camino corto que es el primer elemento de la lista
camino_corto = caminos[0]

#Seleccionamos el camino largo que coincide con el ultimo elemento de la lista
camino_largo = caminos[-1]

print(camino_corto)
print('\n')
print(camino_largo)
print('\n')
print(f'Tras calcular por medio de la funcion shortest simple paths, nos ha generado un total de {len(caminos)} rutas posibles')
print('\n')
print(f'La ruta mas larga segun la funcion shortest simple paths transcurre por {len(camino_largo)} ciudades diferentes')
````

````python
#Iteramos para buscar las coincidencias de los nodos que tienen 2 grados de vecindad

lista = list(G)
ciu_2nds = []
for i in lista:
  if G.degree(i) == 2:
    ciu_2nds.append(i)
  else:
    continue
ciu_2nds

print('Las estaciones que solo tienen 2 vecinos o son de paso son ' + str(', '.join(ciu_2nds)))
````
