a) Queremos hacer un programa que vaya sumando el dinero que gastamos y que, cuando lleguemos a un límite que definimos nosotros, nos avise y detenga el programa. El código debe tener las siguientes características:
- Empezaremos con un dinero inicial 0 y tendremos un límite de 50. <span style = "font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> (**0.25 puntos**)
- En cada iteración, incrementaremos el dinero que hemos gastado por un valor random entre 0 y 10 mediante la función [randint](https://docs.python.org/3/library/random.html#random.randint). <span style = "font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> (**0.5 puntos**)
- En cada iteración, se tiene que mostrar por pantalla cuánto dinero hemos gastado y cuánto nos quedan. Es decir, la diferencia entre el límite y el dinero total gastado en cada iteración. <span style = "font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> (**0.25 puntos**)
- Una vez lleguemos o superemos el límite, se deberá avisar por pantalla y mostrar también cuánto dinero hemos gastado finalmente y cuánto nos hemos pasado del límite. <span style = "font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> (**0.5 puntos**)

(**1.5 puntos**)


````python
import random  #Importamos la libreria random para poder hacer uso de randint 
dinero_inicial = 0 #Definimos la variable "dinero_inicial" que se refiere al computo total del dinero gastado en ese momento
limite = 50  #Definimos la variable "limite" que se refiere al límite que podemos alcanzar con las operaciones 
while limite > 0: #Definimos la operacion de iteracion while para que se repita hasta que la variable "limite" sea menor que 0
  gasto = random.randint(0,10) #Creamos la variable "gasto" para asignarle la cuantificacion del gasto en cada iteraccion por medio de la funcion randint(a,b+1), la cual nos genera un numero entre 0,10 (a=0 y b+1= 10) aleatoriamente 
  dinero_inicial+= gasto #Definimos que por cada iteracion "dinero_inicial"se vera incrementada en relacion al numero aleatorio impuesto en la variable "gasto"
  limite-= gasto #Definimos que por cada iteracion "limite" se vera decrementada en la cantidad aleatoria que haya representado el programa en "gasto"
  if limite > 0: #Dentro del bucle vamos a crear una condicion para que cuando la variable "limite" sea mayor a 0 imprima por pantalla el gasto que llevamos realizado en total por cada iteracion junto con la variable "limite" para expresar cuanto dinero nos podemos seguir gastando
    print(f"lleva gastado usted {dinero_inicial}€ en total, le quedan {limite}€ para seguir gastando ")
  elif limite<= 0: #Definimos otra condicion de cuando el limite sea menor o igual a 0 en el ultimo recorrido del bucle salgamos. 
    break
if limite == 0: #Fuera del bucle seguimos el código definiendo una condicion para que cuando la variable "limite" sea igual a 0 nos indique que el ultimo gasto realizado ha llegado al limite de nuestro presupuesto. No se indica el dinero total gastado debido a que es igual al límite.
  print(f"Con su ultimo gasto hemos cumplido con nuestro límite impuesto de {dinero_inicial}€")
else: #LLegados a este punto, la variable limite siempre va a ser menor o igual a 0, con esta condicion decimos que para cualquier valor negativo de "limite" imprima el valor total gastado y en cuántos euros nos hemos sobrepasado del limite impuesto
  print(f"Usted ha gastado en total {dinero_inicial}€ y se ha pasado del límite {abs(limite)}€")


#Bibliografia utilizada los recursos del aula, y para la funcion abs(): https://www.programiz.com/python-programming/methods/built-in/abs
````

````python
#a2)

list_elem2 = [10, 45, 57, 4, 2, 13, 98, 7]
new_list = []

for x in list_elem2:
    
    if x >= 45: 
        new_list.append(x+1)
    else: 
        new_list.append(x+5)

print(new_list)
````

````python
new_list = [x+1 if x >= 45 else x+5 for x in list_elem2 ] #Se crea la variable lista y se aplica primero la operacion y luego la condicion, finalmente se referencia la lista de donde provienen los elementos que vamos a modificar
print (new_list)
````

````python
#a3)

names = ['Alex','Peter','Anna','John','Felix','Pedro']
new_names = []

for name in names: 
    if name.lower().startswith('a'):
        new_names.append(name)

print(new_names)
````

````python
new_names = [x for x in names if x.lower().startswith('a')]  #Se crea la variable lista, se crea el bucle de for para que muestre los nombres que empiecen por "a"
print (new_names)
````
