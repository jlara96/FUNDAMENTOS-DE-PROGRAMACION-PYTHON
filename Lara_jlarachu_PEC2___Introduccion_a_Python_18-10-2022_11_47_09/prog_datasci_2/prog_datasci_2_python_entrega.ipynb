### Ejercicio 3
Cread el string **roxanneENNAXORanneRox** en una única expresión y utilizando solamente la variable m = `roxanne`.

**Nota**: Para resolver este ejercicio tenéis que utilizar las funciones para convertir un string en mayúsculas que podéis consultar en el seguiente [link](https://docs.python.org/3/library/stdtypes.html?highlight=upper#str.upper).
**(1 punto)** $\color{orange}{\text{EG}}$

```python
#En primer lugar definimos la variable que vamos a utilizar para poder llegar al resultado solo con ella:
m='roxanne'

#Accedemos a la cadena de caracteres 'm' en orden inversa y en masyusculas 
str1= m[::-1].upper()

#Accedemos a los 4 caracteres ultimos de la cadena 'm'
str2= m[3:]

#Accedemos al primer caracter de la cadena de caracteres 'm' y la transformamos en mayuscula
str3=m[0].upper()

#Accedemos a los caracateres comprendidos entre la 2 y 3 posicion de la cadena 'm'
str4=m[1:3]

#Concatenamos los bloques anteriores
resultado = m+str1+str2+str3+str4

#En ultimo lugar escribimos el código para escribir la palabra en negrita e imprimimos el resultado
resultado= '\033[1m'+resultado+'\033[0m'
print(resultado)

# Para convertir el texto en negrita he buscado el codigo en la siguiente página: https://www.delftstack.com/es/howto/python/python-bold-text/
```

**(1.5 punts)** $\color{green}{\text{NM}}$

```python
#En primer lugar importamos la libreria numpy
import numpy as np

#definimos la variable lista con los 5 primeros numeros de la serie "lost"
lost =[23,9,16,15,8]

#imprimimos la suma de todos los numeros de la lista con la funcion de la libreria numpy (np.sum)
print ('a) SUMA:', np.sum(lost))

#imprimos el maximo y el minimo de la variable con la funcion de la libreria numpy (np.max y np.min)
print ('b) MAXIMO:', np.max(lost), '; MINIMO:', np.min(lost))

#imprimimos la desviacion estandar de la variable con la funcion de la libreria numpy (np.std) y redondeamos el resultado a dos digitos decimales con round
print ('c) DESVIACION ESTANDAR:', round(np.std(lost),2))

#ordenamos de mayor a menor los numeros de la variable con la funcion de la libreria numppy (np.sort) añadiendole [::-1] para realizar el reverso
print ('d) ORDEN MAYOR A MENOR:', np.sort(lost)[::-1])

#Bibliografia extraida de los recursos de la asignatura y https://www.analyticslane.com/2020/11/23/numpy-como-ordenar-una-matriz-numpy-en-python/#:~:text=Si%20queremos%20ordenar%20los%20valores,opci%C3%B3n%20para%20hacer%20esto%20directamente.
```
