# **NumPy - Ejercicios**

### Ejercicio 1

(a) Crea una matriz 10x10 de números **enteros** aleatorios comprendidos entre 0 y 10. **(0.5 puntos)** $\color{green}{\text{NM}}$

(b) A partir de la matriz anterior, calcula una [matriz simétrica](https://es.wikipedia.org/wiki/Matriz_sim%C3%A9trica). Puedes quizás inspirarte de la siguiente página web: https://stackoverflow.com/questions/10806790/generating-symmetric-matrices-in-numpy. **(0.5 puntos)** $\color{orange}{\text{EG}}$

(c) Verifica mediante el uso de "slices" que la matriz del apartado (b) es realmente simétrica. **(0.5 puntos)** $\color{green}{\text{NM}}$

Muestra por pantalla los resultados de cada uno de estos pasos.

```python
import numpy as np

#EJERCICI0 1 A.
#Creamos una matriz 10x10 de números enteros aleatorios comprendidos entre 0 y 10.
matrix = np.random.randint(11, size=(10,10))
print(matrix)
print("\n")

#EJERCICIO B
#Creamos la matriz simetrica segun su definicion.
matrix_symm = (matrix + matrix.T)/2
print(matrix_symm)
print("\n")

#EJERCICIO C
#Para comprobar si la matriz es exactamente simetrica recorremos las posiciones de la matriz, continuadamente definimos
#comparando los valores de una posicion con la traspuesta de esa posicion.

for i in range(len(matrix_symm)):
  for j in range(len(matrix_symm)):
    if (matrix_symm[i][j] != matrix_symm[j][i]):
      print("False")
      break 
      #Rompemos el bucle ya que solo necesitamos el primer indicio para afirmar que la 
      #matriz no es simetrica.
    else:
      continue
print("Si existe un mensaje de false anterior, la matriz no es simétrica. En caso contrario, la matriz es simétrica")

#Bibliografia consultada de los recursos de la asignatura y los enlaces proporcionados en el enunciado.
```

### Ejercicio 2

Dadas dos matrices A y B, en este ejercicio vamos a calcular lo siguiente: $tr(A·B^T)$, donde $tr(X)$ indica la [traza de una matriz](https://es.wikipedia.org/wiki/Traza_(%C3%A1lgebra_lineal)) $X$, $X^T$ indica la [transpuesta](https://es.wikipedia.org/wiki/Matriz_transpuesta) de la matriz $X$ y el punto $·$ corresponde al producto entre matrices.

(a) Escoge cualquiera de las matrices generadas en el ejercicio anterior y modifica sus dimensiones para obtener una matriz **20x5**. Ésta será nuestra matriz **A**. **(0.5 points)** $\color{green}{\text{NM}}$

(b) Crea una matriz **20x5** cuyos elementos sean todos igual a 1. Ésta será nuestra matriz **B**. **(0.5 points)** $\color{green}{\text{NM}}$

(c) Con las matrices A y B de los apartados anteriores, calcula la traza de **A** multiplicada por **B transpuesta** mediante un bucle *for* que recorra la matriz resultado del producto. Para resolver este apartado pueden ser de utilidad las funciones [dot](https://numpy.org/doc/stable/reference/generated/numpy.dot.html) y [transpose](https://numpy.org/doc/stable/reference/generated/numpy.transpose.html). **(0.5 points)** $\color{orange}{\text{EG}}$

(d) Calcula de nuevo la traza de **A** multiplicada por **B transpuesta** de nuevo pero, esta vez, utilizando la función adecuada de la librería NumPy.  **(0.5 points)** $\color{red}{\text{EI}}$

Muestra por pantalla los resultados de cada uno de estos pasos.

```python
#EJERCICIO A
#Escogemos la matriz que se crea automaticamente para que siempre podamos comprobar si el código
#funciona correctamente.
matr_a = matrix.reshape(20,5)
print(matr_a)
print("\n")

#EJERCICIO B
#Creamos otra matriz estática con valores 1 con la siguiente función.
matr_b = np.ones([20,5])
print(matr_b)
print("\n")

#EJERCICIO C
#Creamos una matriz nueva multiplicandp las matrices anteriores utilizando dot y previamente
#realizando la transposicion de la matriz_b.
matr_c = np.dot(matr_a, np.transpose(matr_b))
print(matr_c)
#Recorremos las columnas de la nueva matriz indicando que cuando coincida la posición de la columna
#y la fila se sumen.
suma=0
for i in range(len(matr_c)):
  suma+= matr_c[i][i]
print("\n")
print("La traza de la nueva matriz es {}".format(suma))
print("\n")

#EJERCICIO D 
nuevo = np.trace(np.transpose(matr_c))
print("La traza de la nueva matriz calculada con la funcion trace es {}".format(nuevo))

#Bibliografia consultada de los recursos de la asignatura, los enlaces proporcionados en el enunciado.
#TRACE = https://github.com/platzi/algebra-lineal-python/blob/master/08%20-%20Otras%20Funciones%20de%20Algebra%20Lineal/La%20traza%20y%20el%20determinante.ipynb

```

### Ejercicio 4
Calcula la [media ponderada](https://es.wikipedia.org/wiki/Media_ponderada) de la duración de los viajes para el año 2021, cogiendo como *pesos* el número de turistas mensuales de ese mismo año.

**(1 punto)** $\color{red}{\text{EI}}$

```python
#CREAMOS UN DF CON LOS DATOS QUE SEAN DE 2021 Y LA DURACION MEDIA DE LPS VIAJES.
df = data[(data["Concepto turístico"]== "Duración media de los viajes") & (data.año == 2021)]

#CREAMOS OTRO DF NUEVO CON LOS DATOS DE 2021 Y QUE SEAN DE TURISTA. LUEGO AÑADIMOS UNA COLUMNA NUEVA CON
#CON EL PESO DE LA CANTIDAD TOTAL DE 2021 SEGUN EL MES.
df_turistas = data[(data["Concepto turístico"]== "Turistas") & (data.año == 2021)]
df_turistas["peso"] =  df_turistas["Total"]/(df_turistas["Total"].sum())

#REDEFINIMOS EL NOMBRE DE LA COLUMNAS PARA QUE NO SE SUPERPONGAN Y REALIZAMOS EL INNER JOIN
#PARA JUNTAR LOS DF.
df_tur = df_turistas.rename({"Concepto turístico":"definicion","Total":"tot_turist"}, axis = "columns")
df_total = pd.merge(df,df_tur,on=["Periodo"],how="inner")

#DEFINIMOS LAS VARIABLES PARA REALIZAR LA MEDIA PONDERADA.
D = df_total['Total']
W= df_total["peso"]
mp = (D*W).sum()
print(mp)

#Bibliografia consultada de los recursos de la asignatura y los enlaces proporcionados en el enunciado.
#Para utilizar el inner join: https://analisisydecision.es/tipos-de-uniones-join-de-tablas-con-python-pandas/
```

### Ejercicio 7

(a) Utiliza la función [gaussian_kde](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gaussian_kde.html) de la librería `stats` de `scipy` para estimar la densidad de probabilidad correspondiente a la `duración media de los viajes` (todos los años incluidos) y compárala en un mismo gráfico con el histograma del ejercicio 7. **(0.5 puntos)** $\color{orange}{\text{EG}}$

(b) Utiliza la función [quad](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.quad.html) de la librería `integrate` de `scipy` para verificar que la integral de la función de densidad de probabilidad calculada en el apartado anterior es aproximadamente 1. **(Opcional)** $\color{orange}{\text{EG}}$

```python

#EJERCICIO A 
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

#Definimos los datos a los que vamos a calcular kde
x_dfh =(df_hist["Total"])
kde = stats.gaussian_kde(x_dfh)

#Creamos un array con los numeros comprendidos entre los existentes entre el min y max de 
#df_total y tamaño de la columna (37) e incluimos su kde en función del anterior
x_est = np.linspace(6,14,37)
kde_est = kde(x_est)

#Representamos los datos que hemos calculado anteriormente en una misma gráfica
plt.hist(x_dfh, bins=10, density=True)
plt.plot(x_est, kde_est, label="kde estimada", \
 color="r")
plt.title("histograma de duracion media de los viajes")
plt.xlabel("valores")
plt.ylabel("Probabilidad")
plt.legend()
plt.show()

```

```python
#EJERCICIO B
import numpy as np
import scipy
from scipy.integrate import quad

#Realizamos la comprobacion para verificar que la integral de la función de densidad 
#de probabilidad calculada en el apartado anterior es aproximadamente 1
densidad,error = quad(kde, 0,14)
print(f"La integral de la función de densidad anterior es {densidad} y el error cometido es de {error}")


#Bibliografia consultada de los recursos de la asignatura y los enlaces proporcionados en el enunciado.
```
